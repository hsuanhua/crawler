```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
```

```python
options = Options()
options.add_argument("--disable-notifications")

# 使用偽造的 user-agent
ua = "Mozilla/5.0 (Windows NT 10.0; WOW64; rv:53.0) Gecko/20100101 Firefox/53.0"
options.add_argument("user-agent={}".format(ua))         

driver = webdriver.Chrome("D:/自己研究資料/NLP/chromedriver",chrome_options=options)

# 更改網址以前往不同網頁
driver.get("https://www.rottentomatoes.com/m/american_made_2017/reviews") 
```

```python
def crawler(count,total):
    while count < total:
        content = driver.find_elements_by_class_name('the_review')
        for i in range (len(content)):
            date = driver.find_element_by_xpath('//*[@id="content"]/div/div/div/div[2]/div['+str(i+1)+']/div[2]/div[2]/div[1]')
#             //*[@id="content"]/div/div/div/div[2]/div[1]/div[2]/div[2]/div[2]/div[1]
#             //*[@id="content"]/div/div/div/div[2]/div[2]/div[2]/div[2]/div[2]/div[1]
#             //*[@id="content"]/div/div/div/div[2]/div[3]/div[2]/div[2]/div[2]/div[1]
            print(i," ",date.text," ",content[i].text)
            content_list.append(content[i].text)
            date_list.append(date.text)

        count = count + 1
        # 定位搜尋框 點擊按紐
        button = driver.find_element_by_xpath('//*[@id="content"]/div/div/div/nav[2]/button[2]')
#         //*[@id="content"]/div/div/div/nav[2]/button[2]
        button.click()
        time.sleep(5)
        
    return content_list,date_list
```

```python
content_list = []
date_list = []
total = 271/20
count = 0
crawler(count,total)
```

```python
import csv
import pandas as pd
```

```python
df = pd.DataFrame({'date':date_list , 'content': content_list})

df.to_csv('2017 60 專家評論.csv',encoding='utf-8-sig')
```
